# Finviz

```python
import pandas as pd
import numpy as np
from bs4 import BeautifulSoup as soup
from urllib.request import Request, urlopen
```

```python
# Input
symbol = "AAPL"

# Set up scraper
url = ("http://finviz.com/quote.ashx?t=" + symbol.lower())
req = Request(url, headers={'User-Agent': 'Mozilla/5.0'})
webpage = urlopen(req).read()
html = soup(webpage, "html.parser")
```

```python
def get_news():
    try:
        # Find news table
        news = pd.read_html(str(html), attrs = {'class': 'fullview-news-outer'})[0]
        links = []
        for a in html.find_all('a', class_="tab-link-news"):
            links.append(a['href'])

        news["Ticker"] = symbol
        news['Link'] = links
        
        # Clean up news dataframe
        news.columns = ['Date', 'Title', 'Ticker', 'Link']
        return news

    except Exception as e:
        return e

news = get_news()
news.head(5)
```

# MarketWatch

```python
ticker = "AAPL"
url = f"https://www.marketwatch.com/investing/stock/{ticker.lower()}?mod=search_symbol"
request = requests.get(url)
parser = BeautifulSoup(request.text, "html.parser")

news_html = parser.find_all('div', {'class': 'article__content'})
```

```python
import re

def clean_text(text):
    text = " ".join(re.split("\s+", text, flags=re.UNICODE))
    text = re.sub("^\s+|\s+$", "", text, flags=re.UNICODE)

    return text
```

```python
sentiments = []
for i in range(0, len(news_html)):
    try:
        date = news_html[i].find('span', {'class': 'article__timestamp'}).text
        title = clean_text(news_html[i].find('a', {'class': 'link'}).text)
        link = news_html[i].find('a', {'class': 'link'})["href"]
    except: 
        continue

    sentiments.append(
            {
                'Ticker': ticker,
                'Date': date,
                'Title': title,
                'Link': link
            }
        )
sentiments
```
